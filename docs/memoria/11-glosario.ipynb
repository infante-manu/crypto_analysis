## 11. Glosario 
### Explicación de los Resultados del Backtest

El análisis del rendimiento del backtest se realiza mediante un conjunto de métricas clave, cada una de las cuales ayuda a evaluar la efectividad y el riesgo de la estrategia de trading. A continuación se detallan las métricas calculadas:

#### 1. **'Initial Capital'**
   - **Descripción:** Capital inicial con el que se comienza el backtest.
   - **Cálculo:** Es el valor definido al iniciar el backtest, sin cambios durante la ejecución.

#### 2. **'Final Capital'**
   - **Descripción:** Capital final tras ejecutar todas las operaciones de compra y venta.
   - **Cálculo:**  
     $$ \text{Capital Final} = \text{Capital Inicial} + \text{Ganancias/Pérdidas Totales} $$

#### 3. **'Total Trades'**
   - **Descripción:** Número total de operaciones realizadas durante el backtest.
   - **Cálculo:** Es simplemente el número de elementos en la lista `self.trades`, que contiene todas las operaciones de compra y venta.

#### 4. **'Winning Trades'**
   - **Descripción:** Número de operaciones que resultaron en ganancias.
   - **Cálculo:**  
     $$ \text{Operaciones Ganadoras} = \text{Número de operaciones donde} \, \text{Profit} > 0 $$

#### 5. **'Losing Trades'**
   - **Descripción:** Número de operaciones que resultaron en pérdidas.
   - **Cálculo:**  
     $$ \text{Operaciones Perdedoras} = \text{Número de operaciones donde} \, \text{Profit} < 0 $$

#### 6. **'Total Profit'**
   - **Descripción:** Beneficio total obtenido o pérdida total durante el backtest.
   - **Cálculo:**  
     $$ \text{Beneficio Total} = \text{Capital Final} - \text{Capital Inicial} $$

#### 7. **'Sharpe Ratio'**
   - **Descripción:** El Sharpe Ratio mide el rendimiento ajustado por riesgo. Un ratio mayor indica una mejor relación entre retorno y volatilidad.
   - **Fórmula:**
     $$ \text{Sharpe Ratio} = \frac{\text{Media de los retornos}}{\text{Desviación estándar de los retornos}} \times \sqrt{252} $$

     El Sharpe Ratio ajusta el retorno para tener en cuenta la volatilidad del portafolio, utilizando una base anual de 252 días de negociación. Un Sharpe Ratio más alto indica una estrategia con mejores rendimientos ajustados por riesgo.
   - **Referencia:** [Sharpe Ratio - Wikipedia](https://en.wikipedia.org/wiki/Sharpe_ratio)

#### 8. **'Annual Return (%)'**
   - **Descripción:** El Retorno Anual refleja la rentabilidad anualizada del portafolio como porcentaje.
   - **Fórmula:**
     $$ \text{Retorno Anual} = \left( \frac{\text{Capital Final}}{\text{Capital Inicial}} - 1 \right) \times 100 $$

     Este valor se calcula tomando el valor final del portafolio y comparándolo con el capital inicial.
   - **Referencia:** [Annualized Return - Wikipedia](https://en.wikipedia.org/wiki/Annualized_return)

#### 9. **'Max Drawdown (%)'**
   - **Descripción:** El Máximo Drawdown muestra la mayor pérdida desde el pico más alto hasta el punto más bajo del portafolio durante el período del backtest.
   - **Fórmula:**
     $$ \text{Max Drawdown} = \frac{\text{Valor Máximo} - \text{Valor Mínimo Posterior}}{\text{Valor Máximo}} \times 100 $$

     El Máximo Drawdown es una métrica crucial para evaluar el riesgo, ya que muestra la mayor caída desde el valor más alto alcanzado hasta el valor más bajo posterior.
   - **Referencia:** [Maximum Drawdown - Wikipedia](https://en.wikipedia.org/wiki/Maximum_drawdown)

### Bandas de Bollinger

Las Bandas de Bollinger son un indicador técnico que mide la volatilidad del precio y establece un rango en torno a una media móvil. Se componen de tres líneas:

1. **Línea Superior (Upper Band):**  
   $$ \text{Upper Band} = \text{Media Móvil} + (K \times \text{Desviación Estándar}) $$

2. **Línea Inferior (Lower Band):**  
   $$ \text{Lower Band} = \text{Media Móvil} - (K \times \text{Desviación Estándar}) $$

3. **Media Móvil (Middle Band):**  
   $$ \text{Middle Band} = \text{Media Móvil Simple de X Períodos} $$

**Interpretación:**
- Cuando el precio se acerca a la **Upper Band**, el activo puede estar sobrecomprado, lo que sugiere una posible venta.
- Cuando el precio se acerca a la **Lower Band**, el activo puede estar sobrevendido, lo que sugiere una posible compra.

El valor de \( K \) es un parámetro que generalmente se establece en 2 para calcular las bandas a dos desviaciones estándar del precio.
- **Referencia:** [Bollinger Bands - Wikipedia](https://en.wikipedia.org/wiki/Bollinger_Bands)

### RSI (Índice de Fuerza Relativa)

El RSI es un indicador de momentum que mide la velocidad y el cambio de los movimientos de precio. Su valor varía entre 0 y 100, y ayuda a identificar condiciones de sobrecompra o sobreventa.

**Fórmula:**
$$ RSI = 100 - \frac{100}{1 + RS} $$

donde:
$$ RS = \frac{\text{Promedio de los cierres alcistas}}{\text{Promedio de los cierres bajistas}} $$

**Interpretación:**
- Un RSI superior a 70 indica que el activo está **sobrecomprado**, lo que podría sugerir una venta.
- Un RSI inferior a 30 indica que el activo está **sobrevendido**, lo que podría sugerir una compra.

Este indicador es útil para identificar puntos potenciales de reversión de tendencia.
- **Referencia:** [Relative Strength Index (RSI) - Wikipedia](https://en.wikipedia.org/wiki/Relative_strength_index)
